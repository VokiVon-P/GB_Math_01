#### 1. Задание
Даны два вектора в трехмерном пространстве: (20,20,20) и (0,0,-20)

Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
A =[5, 10, 15]

sum_a = 0
for _a in A:
    a = _a**2
    sum_a+=a
length_A = np.sqrt(sum_a)    

print("Длинна вектора А ",length_A)
```

#### 5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.


```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = -6*X + 2*Y
Z2 = -6*X + 2*Y +40
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```
